```python
df = pd.read_csv('space.csv', delimiter=';')
```

```python
num_usa = len(df[df['Location'].str.contains('USA')]['Company Name'].unique())
num_chi = len(df[df['Location'].str.contains('China')]['Company Name'].unique())

plt.bar(('USA', 'China'), (num_usa, num_chi))
```

# Exerc√≠cio3

```python
df_filtered_by_roscosmos = df[df['Company Name'] == 'Roscosmos']
roscomos_success_number = len(df_filtered_by_roscosmos[df_filtered_by_roscosmos['Status Mission'] == 'Success'])
roscomos_non_success_number = len(df_filtered_by_roscosmos[df_filtered_by_roscosmos['Status Mission'] != 'Success'])

suc_pct = roscomos_success_number / (roscomos_success_number + roscomos_non_success_number)
non_suc_pct = roscomos_non_success_number / (roscomos_success_number + roscomos_non_success_number)

plt.pie(
    x=[suc_pct, non_suc_pct],
    labels=['sucesso', 'fracasso'],
    autopct='%1.1f%%'
)
```
